```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from scipy.fftpack import fft, ifft, fftshift, fftfreq
```

# Coseno

```python
A = 1
f = 10
phi = np.pi/2*0

c = 2
fs = f*100
t = np.arange(-c/f,c/f+1/fs,1/fs)
M = len(t)

# Seal
x = A * np.cos(2*np.pi*f*t - phi)

# zero phase
h1 = int(np.float(M/2))
h2 = int(np.ceil(M/2))

# Espectro
x_zp = np.append(x[h2:], x[:h1])

X = fftshift(fft(x))
freq = fftshift(fftfreq(X.size, d=1/fs))

# Magnitud y Fase
Xm = np.abs(X)/X.size
#Xm_db = 20*np.log10(Xm/np.max(Xm))
Xp = np.unwrap(np.angle(X, True))


plt.figure(figsize=(7.5,5))

plt.plot(t,x)
plt.grid(True)
#plt.xlim([0,0.2])
plt.title("Coseno")
plt.xlabel("tiempo [seg]")
plt.ylabel("Amplitud")


plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
plt.plot(freq,Xm)
plt.grid(True)
plt.xlim([-50,50])
plt.title("Coseno")
plt.xlabel("Frecuencia [Hz]")
plt.ylabel("Amplitud")


plt.subplot(1,2,2)
plt.plot(freq,Xp)
plt.grid(True)
plt.title("Coseno")
plt.xlabel("Frecuencia [Hz]")
plt.ylabel("Fase[deg]")
```

# Cuadrada

```python
A = 1
f = 10
phi = 0

c = 5
fs = f*100
t = np.arange(-c/f,c/f+1/fs,1/fs)
M = len(t)


# Seal
x = A * np.cos(2*np.pi*f*t - phi)
idx_gt = np.where(x>0)
idx_lt = np.where(x<0)
x[idx_gt]=1
x[idx_lt]=0

# zero phase
h1 = int(np.float(M/2))
h2 = int(np.ceil(M/2))

# Espectro
x_zp = np.append(x[h2:], x[:h1])

X = fftshift(fft(x))
freq = fftshift(fftfreq(X.size, d=1/fs))

# Magnitud y Fase
Xm = np.abs(X)/X.size
#Xm_db = 20*np.log10(Xm/np.max(Xm))
Xp = np.unwrap(np.angle(X, True))


plt.figure(figsize=(7.5,5))

plt.plot(t,x)
plt.grid(True)
#plt.xlim([0,0.2])
plt.title("Cuadrada")
plt.xlabel("tiempo [seg]")
plt.ylabel("Amplitud")


plt.figure(figsize=(7.5,5))
#plt.subplot(1,2,1)
plt.plot(freq,Xm)
plt.grid(True)
plt.xlim([-250,250])
plt.title("Modulo")
plt.xlabel("Frecuencia [Hz]")
plt.ylabel("Amplitud")


#plt.subplot(1,2,2)
#plt.plot(freq,Xp)
#plt.grid(True)
#plt.title("Fase")
#plt.xlabel("Frecuencia [Hz]")
#plt.ylabel("Fase[deg]")
```
